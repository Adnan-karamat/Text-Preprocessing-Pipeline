```python
dataset = [
    "I feel sooo sad these days!!! ðŸ˜¢ðŸ˜¢ Nothing seems right... I can't sleep properly. Life is meaningless. Visit http://sadlife.com",

    "Why am I always anxious?? My heart races all the time!!! I overthink everything. I'm tired of this feeling. ðŸ˜ž #stress",

    "I don't want to live anymore... everything hurts. Nobody understands me. I feel empty inside. Maybe I should disappear.",

    "Some days are okay, but other days are unbearable!!! Mood swings are killing me ðŸ˜­ðŸ˜­ I hate this pain.",

    "I am fine today, just a little tired from work. Life is busy but manageable. Hoping for better days ahead."
]
```

```python
for text in dataset:
    print(text)
```

```python
import re

def remove_urls(text):
    return re.sub(r'http\S+|www\S+', '', text)
```

```python
dataset_no_urls = [remove_urls(text) for text in dataset]
```

```python
for text in dataset_no_urls:
    print(text)
```

```python
def remove_emojis(text):
    return re.sub(r'[^\w\s,.!?]', '', text)
```

```python
dataset_no_emojis = [remove_emojis(text) for text in dataset_no_urls]
```

```python
for text in dataset_no_emojis:
    print(text)
```

```python
def to_lowercase(text):
    return text.lower()
```

```python
dataset_lower = [to_lowercase(text) for text in dataset_no_emojis]
```

```python
for text in dataset_lower:
    print(text)
```

```python
def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)
```

```python
dataset_clean_chars = [remove_special_characters(text) for text in dataset_lower]
```

```python
for text in dataset_clean_chars:
    print(text)
```

```python
def remove_extra_spaces(text):
    return ' '.join(text.split())
```

```python
dataset_no_spaces = [remove_extra_spaces(text) for text in dataset_clean_chars]
```

```python
for text in dataset_no_spaces:
    print(text)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in stop_words])
```

```python
dataset_no_stopwords = [remove_stopwords(text) for text in dataset_no_spaces]
```

```python
for text in dataset_no_stopwords:
    print(text)
```
